### 1.2 Изучите стоимости недвижимости

```python
df.head()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['price'], bins = 50)
plt.xlabel('Стоимость, млн')
plt.ylabel('Кол-во домов')
plt.title('Стоимость недвижимости');
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['price'])
plt.xlabel('Стоимость, млн')
plt.title('Стоимость недвижимости');
```

### Вывод:
- Стоимость 50% домов от 0,4 млн до 0,7 млн; 
- Есть сильные выбросы в стоимости домов, максимальная стоимость ближе к 8 млн;
- Минимальная стоимость дома 0,1 млн.;
- У более 6000 домов стоимость недвижимости составляет от 0,2 до 0,4 млн. 

###  1.3 Изучите распределение квадратуры жилой площади

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['sqft_living'], bins = 50)
plt.xlabel('Площадь')
plt.ylabel('Кол-во домов')
plt.title('Квадратура жилой площади');
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['sqft_living'])
plt.xlabel('Площадь')
plt.title('Квадратура жилой площади');
```

### Вывод:
- У 50% домов площадь составялет от 1500 до 2500;
- Максимальная площадь составляет примерно 13800;
- Больше всех домов с площадью 1600;
- Минимальная площадь составляет 400.

### 1.4 Изучите распределение года постройки

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['yr_built'], bins = 50)
plt.xlabel('Год')
plt.ylabel('Кол-во')
plt.title('Год постройки');
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['yr_built'])
plt.xlabel('Год')
plt.title('Год постройки');
```

### Вывод:
- Самая рянняя постройка дома была в 1900 году;
- В 2007-2009 годах было маскимальное кол-во построек; 
- 50% домов были построены с 1950 по 1998 года;
- Последние дома были построены в 2015 году; 
- В 1932-1933 годах было наименьшее число построек.

### 2.2 Изучите распределение этажей домов

```python
df.head()
```

```python
data_2 = df['floors'].value_counts()
```

```python
plt.figure(figsize=(6, 4))
plt.pie(data_2, autopct='%1.1f%%', labels=data_2.index)
plt.title('Количество этажей');
```

```python
df['floors'].value_counts().plot.bar()

plt.xlabel('Этажи')
plt.ylabel('Кол-во домов')
plt.title('Количество этажей')
plt.xticks (rotation = 0);
```

### Вывод:
- Почти половина домов с одним этажом(49,4%);
- Меньше всего домов с 3.5 этажами;
- С 2-мя этажами количество домов занимает 38,1%.

### 2.3 Изучите распределение состояния домов

```python
data_3 = df['condition'].value_counts()
plt.figure(figsize=(6, 4))
plt.pie(data_3, autopct='%1.1f%%', labels=data_3.index)
plt.title('Состояние домов');
```

```python
df['condition'].value_counts().plot.bar()

plt.xlabel('Состояние')
plt.ylabel('Кол-во домов')
plt.title('Состояние домов')
plt.xticks (rotation = 0);
```

### Вывод:
- Больше половины домов(64,9%) с оценкой состояния 3;
- Меньше всех по количеству домов с оценкой состояния 1; 
- Дома с отличным состоянием занимают долю 7,9%;
- С оценкой состояния 4 - 26,3% домов.
